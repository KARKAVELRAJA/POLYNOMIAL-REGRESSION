## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("Position_Salaries.csv")
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.iloc[:,1:2].values
print(input)
output = data.iloc[:,-1].values
print(output)
```

## ***VISUALIZING THE DATASET***

```python
import matplotlib.pyplot as plt
plt.xlabel("POSITION_LEVEL")
plt.ylabel("SALARY")
plt.scatter(input,output)
```

## ***POLYNOMIAL FEATURE SCALING***

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_scaling = PolynomialFeatures(degree=4)
polynomial_input   = polynomial_scaling.fit_transform(input)

polynomial_input
```

## ***TRAINING THE DATASET***

```python
from sklearn.linear_model import LinearRegression

Polynomial_Regression_Model = LinearRegression()

Polynomial_Regression_Model.fit(polynomial_input,output)
```

## ***PREDICTING THE OUTPUT USING THE TRAINED MODEL***

```python
Y = 5

X = polynomial_scaling.fit_transform(np.reshape(Y,(-1,1)))

print("SALARY FOR THE POSITION_LEVEL",Y,"IS",Polynomial_Regression_Model.predict(X))
```

## ***VISUALIZING THE RESULT OF THE TRAINED MODEL***

```python
import matplotlib.pyplot as plt
plt.xlabel("POSITION_LEVEL")
plt.ylabel("SALARY")
plt.scatter(input,output)
plt.plot(input,Polynomial_Regression_Model.predict(polynomial_scaling.fit_transform(input)),color="red")
```
